```python
import pandas as pd

df_test = pd.read_csv('https://raw.githubusercontent.com/bartowen/desicion_tree_algo/refs/heads/main/data/processed/clean_test.csv')
df_train = pd.read_csv('https://raw.githubusercontent.com/bartowen/desicion_tree_algo/refs/heads/main/data/processed/clean_train.csv')

df_test.head()
```

```python
df_test.head()
```

```python
X_train = df_train.drop(["Outcome"], axis = 1)
y_train = df_train["Outcome"]
X_test = df_test.drop(["Outcome"], axis = 1)
y_test = df_test["Outcome"]
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# Grilla

from sklearn.model_selection import GridSearchCV


# Hiperparametros

grid = {
    'n_estimators': [50, 150, 200],
    'learning_rate': [0.01, 0.05],
    'max_depth': [3, 7, 10],
    'min_samples_split': [2, 7, 9],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(model,grid,scoring= 'accuracy', cv=5)

grid_search.fit(X_train,y_train)
```

```python
# Cantidad de modelos en la grilla

# es una practica común en multiplicaciones acumulativas pq el valor es nuetro para la multiplicación. n*1 no cambia el resultado

num_models = 1

# grid.items() -> este metodo devuelve una vista de los pares clave-valor en el diccionario. Por ejemplo, para el primer elemento, param sería 'n_estimators' y value sería [50,150,200]
# len(values) -> devuelve la cantidad de elementos en la lista de opciones para el hiperparámetro actual. 'n_estimatos' = 3


for param, values in grid.items():
    num_models *= len(values) 

print(f'Total de modelos a ejecutar:', num_models)
```
